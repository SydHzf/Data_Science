```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = "https://www.espncricinfo.com/rankings/content/page/211271.html"
response = requests.get(base_site)
response.status_code
response.ok
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
all_tables = soup.find_all("table",{"class":"StoryengineTable"})
```

```python
ranking_dict = {}
h3 = [i.text for i in soup.find_all("h3")]

for j in range(len(all_tables)):
    table = all_tables[j]
    tr = table.find_all("tr")    
    
    data = []

    for i in tr:
        if i.find_all("th") != []:
            data.append([i.text for i in i.find_all("th")])
        elif i.find_all("td") != []:
            data.append([i.text for i in i.find_all("td")])
    
    ranking = pd.DataFrame(data[1:], columns = data[0])
    
    ranking_dict[h3[j]] = ranking
#     print(h3[j])
#     print(ranking)
```

```python
ranking_dict.keys()
```

```python
ranking_dict["ICC Women's T20 Team Rankings"]
```

```python
writer = pd.ExcelWriter('ranking.xlsx', engine='xlsxwriter')

for key,value in ranking_dict.items():
    value.to_excel(writer,sheet_name=key, index=False)
    
writer.save()
```
