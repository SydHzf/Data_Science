```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = "https://www.espncricinfo.com/rankings/content/page/211271.html"
response = requests.get(base_site)
response.ok
```

```python
html = response.text
soup = BeautifulSoup(html,"html.parser")
soup
```

```python
all_table = soup.find_all("table",class_ = "StoryengineTable")
all_table
```

```python
test = all_table[0]
```

```python
tr = test.find_all("tr")
tr
```

```python
test_data = []
for i in tr:
    if i.find_all("th") != []:
#         print(i.find_all("th"))
         test_data.append([i.text for i in i.find_all("th")])  
    elif i.find_all("td") != []:
#         print(i.find_all("td"))
        test_data.append([i.text for i in i.find_all("td")])
```

```python
test_data
```

```python
import pandas as pd
```

```python
pd.DataFrame(test_data[1:],columns = test_data[0])
```

```python
ranking_dict = {}
h3 = [i.text for i in soup.find_all("h3")]
for j in range(len(all_table)):
    table = all_table[j]
    tr = table.find_all("tr")
    data = []
    
    for i in tr:
        if i.find_all("th") != []:
    #         print(i.find_all("th"))
             data.append([i.text for i in i.find_all("th")])  
        elif i.find_all("td") != []:
    #         print(i.find_all("td"))
            data.append([i.text for i in i.find_all("td")])

    ranking = pd.DataFrame(data[1:], columns = data[0])
    ranking_dict[h3[j]] = ranking 
```

```python
ranking_dict
```

```python
ranking_dict["ICC Twenty20 Rankings"]
```

```python
writer = pd.ExcelWriter('ranking.xlsx',engine='xlsxwriter')

for key,value in ranking_dict.items():
    value.to_excel(writer,sheet_name = key,index = False)
writer.save()
```
