```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/2/"
response = requests.get(base_site)
response.status_code
response.ok
```

```python
html = response.text
soup = BeautifulSoup(html, "lxml")
```

```python
divs = soup.find_all("div",{"class":"col-sm-18 col-full-xs countdown-item-content"})
divs
```

```python
headings = [div.find("h2") for div in divs]
```

```python
movie_name = [i.find("a").text for i in headings]
```

```python
headings[0].find("span",class_ = "start-year")
```

```python
year = [i.find("span",class_ = "start-year").text.strip("()") for i in headings]
```
