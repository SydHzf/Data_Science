```python
data = []
for i in links[:15]:
    link = i.get("href")
    #print(link)
    base_sites = link
    responses = requests.get(base_sites)
    text = responses.text
    soupy = BeautifulSoup(text,"lxml")
    table = soupy.find_all("table", class_ = "resp_table currency_table")
    ta = table[0].find_all("td")
    test = {}
    a = 0
    head = []
    for i in table[0].find_all("th"):
        head.append(i.text)
    for j in ta:
        test[head[a]] = j.text
        a += 1
#         test.append(j.text)
    data.append(test)
data    
```
