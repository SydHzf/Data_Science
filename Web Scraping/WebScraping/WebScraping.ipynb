# Rotten Tomatoes

```python
base_site = "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/2/"
response = requests.get(base_site)
response

response.status_code

response.ok

html = response.text
html
```

```python
soups = BeautifulSoup(html,"html.parser")
```

```python
content_body = soups.find("div" ,class_ = "articleContentBody")
content_body
```

```python
blocks = content_body.find_all("div",class_ = "row countdown-item")
blocks,len(blocks)
```

```python
jurassic = blocks[0]
```

```python
jurassic
```

```python
row = []

h2 = jurassic.find("h2")
name = h2.find("a").text
year = h2.find("span",class_ = "subtle start-year").text
MeterScore = h2.find("span",class_ = "tMeterScore").text

main_div = jurassic.find("div",class_="col-sm-24")
consensus = main_div.find("div",class_="info critics-consensus").text.lstrip("Critics Consensus:")
cast = main_div.find("div",class_="info cast").text.lstrip("\nStarring:")
director = main_div.find("div",class_="info director").text.lstrip("\nDirected By:")

row.append([name,year,MeterScore,consensus,cast,director])
```

```python
row
```

```python
table = []
for block in blocks:
    h2 = block.find("h2")
    name = h2.find("a").text
    year = h2.find("span",class_ = "subtle start-year").text
    MeterScore = h2.find("span",class_ = "tMeterScore").text

    main_div = block.find("div",class_="col-sm-24")
    consensus = main_div.find("div",class_="info critics-consensus").text.lstrip("Critics Consensus:")
    cast = main_div.find("div",class_="info cast").text.lstrip("\nStarring:")
    director = main_div.find("div",class_="info director").text.lstrip("\nDirected By:")

    table.append([name,year,MeterScore,consensus,cast,director])
```

```python
table
```

```python
pd.DataFrame(table,columns=["Name","Year","TMeter","Consensus","Cast","Director"] )
```

# urdu Point

```python
base_link = "https://www.urdupoint.com/business/directory/karachi/182/leather-products.html"
response = requests.get(base_link)
```

```python
response.ok
```

```python
html = response.text
html = BeautifulSoup(html,"lxml")
# html
```

```python
div = html.find("div",{'id' : "list_items"})
```

```python
anchor = div.find_all("a",class_ = "cat_style_list")
```

```python
links = [i.get("href") for i in anchor]
links = links[:10]
```

```python
len(links),links[0]
```

```python
df = []
phli = links[0]
response = requests.get(phli)

html = response.text
html = BeautifulSoup(html,"lxml")
table = html.find("table",class_="resp_table currency_table")
td = [i.text for i in table.find_all("td")]
df.append(td)
```

```python
df
```

```python
df = []
for link in links:
    
    response = requests.get(link)

    html = response.text
    html = BeautifulSoup(html,"lxml")
    table = html.find("table",class_="resp_table currency_table")
    td = [i.text for i in table.find_all("td")]
    th = [i.text for i in table.find_all("th")]
    hd = {k:v for k,v in zip(th,td)}
    df.append(hd)
```

```python
df
```

```python
dframe = pd.DataFrame(df)
dframe
```
