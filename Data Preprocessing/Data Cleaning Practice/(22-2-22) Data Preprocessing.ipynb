```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
import seaborn as sns
import warnings
pd.set_option('display.float_format', lambda x : '%.30f' % x)
pd.options.display.max_columns = None
pd.options.display.float_format = "{:,.3f}".format
warnings.filterwarnings('ignore')
sns.set()
```

```python
dataset  = pd.read_excel("A3 general ledger all entries.xlsx")
```

```python
dataset = dataset.iloc[:-25,:]
```

```python
# Deleting Account Total
# dataset.loc[dataset.loc[dataset.Account.str.startswith("Account Total").notnull()].Account.str.startswith("Account Total")]
totals_ind = dataset.loc[dataset.Account.str.startswith("Account Total") == True].index
```

```python
dataset.drop(totals_ind,inplace = True)
# dataset.drop(dataset.loc[dataset.Account.str.startswith("Transaction") == True].index)
headings = list(dataset.iloc[1])
```

```python
#  Deleting the headinglines
dataset.drop(dataset.loc[dataset.Account.str.startswith("Transaction") == True].index,inplace = True)
```

```python
dataset
```

```python
dataset = dataset.loc[dataset.Account.isnull() != True]
dataset
dataset = dataset.loc[dataset.Description.isnull() != True]
```

```python
acc_no = list(dataset.loc[dataset["Beginning Balance"].isnull() == True,"Account"])
acc_no.__len__()
acc_name = list(dataset.loc[dataset["Beginning Balance"].isnull() == True,"Description"])
acc_name.__len__()
```

```python
dataset["Beginning Balance"].fillna("-")
```

```python
no = []
name = []
count = -1
for i in dataset["Beginning Balance"].fillna("-"):
    if i=="-":
        count+=1# 0100,0110,obj,obj
    else:
        no.append(acc_no[count])
        name.append(acc_name[count])        
```
