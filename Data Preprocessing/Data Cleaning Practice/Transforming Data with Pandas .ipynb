```python
def category(ele):
    if ele>1:
        return "high"
    else:
        return "low"
```

```python
eco_high_low = happiness["Economy"].map(category)
```

```python
happiness[eco_high_low == "high"]
```

```python
def categories(ele,x):
    if ele>x:
        return "high"
    else:
        return "low"
```

```python
happiness["Economy"].apply(categories,x = 1)
```
