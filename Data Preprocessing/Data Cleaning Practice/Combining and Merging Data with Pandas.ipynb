```python
happiness_15['Year'] = "2015"
happiness_16['Year'] = "2016"
happiness_17['Year'] = "2017"
happiness_15.head(2)
```

```python
three_15 = happiness_15[['Country','Happiness Score','Year']]
three_16 = happiness_16[["Year",'Happiness Score','Country']] 
three_17 = happiness_17[["Year",'Happiness.Score','Country']]
```

```python
three_15_16 = pd.concat([three_15,three_16],axis = 0,ignore_index = True)
three_15_16
```

```python
three_17.rename({"Happiness.Score":"Happiness Score"},axis = 1,inplace = True)
```

```python
Happiness = pd.concat([three_15,three_16,three_17],axis = 0,ignore_index = True)
```

```python
Happiness = Happiness.pivot_table(index = "Country" , values = "Happiness Score",aggfunc = 'mean')
Happiness = Happiness.sort_values(by = 'Happiness Score',ascending = False)
len(Happiness)
```

```python
Happiness.insert(0, "Rank", [str(i)+"th" if i > 3 else i for i in range(1,len(Happiness)+1)], allow_duplicates=False)
```
