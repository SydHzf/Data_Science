```python
def predicted_price(new_listing,k = 5):
    new_df = dataset.copy()
    new_df["distance"] = new_df['accommodates'].apply(lambda x : np.sqrt((x - new_listing)**2))
    new_df = new_df[new_df["distance"] == 0]
    np.random.seed(0)
    shuffle_index = np.random.permutation(new_df.index)
    new_df = new_df.loc[shuffle_index]    
    new_df["price"] = new_df["price"].str.strip("$").str.replace(",","").astype(float) 
    neibour = new_df["price"].head(k) 
    mean_price = neibour.mean()
    return mean_price
```
